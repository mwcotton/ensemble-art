Input Generation
---
Generates some simple, artificial inputs for the processing, either as sample inputs, noise maps, etc.

```python
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
```

```python
x = np.arange(-5, 5, 0.02)
y = np.arange(-5, 5, 0.02)
xx, yy = np.meshgrid(x, y, sparse=True)
z = 255*np.exp(-(xx**2 + yy**2)/2)
plt.imshow(z)
Image.fromarray(z.astype(np.uint8)).save('./inputs/central500x500.png')
```

```python
x = np.arange(-10, 10, 0.02)
y = np.arange(-10, 10, 0.02)
xx, yy = np.meshgrid(x, y, sparse=True)
z = 255*np.exp(-(xx**2 + yy**2)/5)
plt.imshow(z)
Image.fromarray(z.astype(np.uint8)).save('./inputs/central1000x1000.png')
```

```python
x = np.arange(-10, 10, 0.02)
y = np.arange(-10, 10, 0.02)
xx, yy = np.meshgrid(x, y)
z=np.zeros(np.shape(xx))
for i in [7.5, 2.5, -2.5, -7.5]:
    for j in [7.5, 2.5, -2.5, -7.5]:
        z += 255*np.exp(-((xx-i)**2 + (yy-j)**2)/1)
plt.imshow(z)
Image.fromarray(z.astype(np.uint8)).save('./inputs/blobs1000x1000.png')
```
